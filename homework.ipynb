### Problem 12 (automatically graded)

Write a function `p12` that takes in a sequence of strings and returns a list of the strings starting with the prefix "input_".
The list should be in the same order as the input sequence.

```python
# YOUR CHANGES HERE

def p12(column_names):
    """
    Takes a sequence of strings (column_names) and returns a list of the 
    strings starting with the prefix "input_". The list should be in the 
    same order as the input sequence.
    """
    
    # Use a list comprehension to iterate through column_names 
    # and select strings that start with "input_".
    filtered_list = [
        name 
        for name in column_names 
        if name.startswith("input_")
    ]
    
    return filtered_list
```

### Problem 19 (automatically graded)

The data frame in `q19` contains the first names of actors and their hometowns.
Set `p19` to be a data frame of pairs of actors with the same hometown.
`p19` should have two columns, `name_1` and `name_2` and the value in `name_1` should before the value in `name_2` to avoid matching an actor with themselves and repeating a pair with both orders.

```python
# DO NOT CHANGE

q19 = pd.DataFrame({"name": ["Mel", "John", "Tom", "Benjamin", "Michael", "Emily", "Tom"],
                    "hometown": ["Peekskill", "Newton", "Concord", "Newton", "Newton", "London", "London"]})
q19
```

```python
# 1. Self-Merge (Inner Join) on 'hometown'
merged_df = q19.merge(
    q19,
    on='hometown',
    suffixes=('_1', '_2')
)

# 2. Filter to get unique pairs
# This filters out self-matches (name_1 != name_2) 
# AND ensures only one version of each pair is kept (e.g., Alice-Bob but not Bob-Alice).
p19_pairs = merged_df[
    (merged_df['name_1'] != merged_df['name_2']) & 
    (merged_df['name_1'] < merged_df['name_2'])
]

# 3. Select and set the final output
# CORRECTION: Only select the 'name_1' and 'name_2' columns.
p19 = p19_pairs[['name_1', 'name_2']]
```
